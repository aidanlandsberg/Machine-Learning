### 2.2 Analysis of the Iris Dataset

We begin the analysis by loading the iris data set through the given resources - namely the load_iris module given in the project resources:  

```python
%matplotlib inline
import load_iris as iris
import numpy as np
import scipy as sc
from matplotlib import pylab as plt
from sklearn.decomposition import PCA

iris_file = ('iris_train.txt')
iris_data = iris.load_iris2(iris_file)
print('No. of elements in data set:')
print(len(iris_data))
```

it is evident from loading the iris data set that there exists 90 instances of the iris flower from the three separate classes, namely Setosa, Versicolour and Virginica. The data presents the four parameters afterwhich the class belongs to is shown. 

We can then proceed to fit the model with training data (in this case the aforementioned iris data set) afterwhich dimensionlity reduction can be applied to project the aforemensioned data set onto the first 2 principal components. This is realised using a principal component analysis algorithm that utilises singular value decomposition to project data onto a lower dimensional space. 

By using the expalined_variance_ratio attribute in the PCA class, it can be deduced that the majority of the variance is indeed explained within the first two proncipal components. The functions outout - shown below 

```python
pca = PCA(n_components=2)
new_iris_data = pca.fit_transform(iris_data)
print('Variance ratio amongst the first two principal compinents:')
print(pca.explained_variance_ratio_)
print('Ultametly, it is proven that above 97% of the variance is explained amongst the first 2 components.')
```

```python
#matplotlib.rcParams.update({'font.size': 18, 'text.usetex': True})
n = len(new_iris_data)
x = np.linspace(0.,n,n)
fig, ax = plt.subplots()
ax.plot(x, iris_data[:,0], label = "Component 1")
ax.plot(x, iris_data[:,1], label = "Component 2")
ax.plot(x, iris_data[:,2], label = "Component 3")
ax.plot(x, iris_data[:,3], label = "Component 4")
ax.set_xlabel('Instances')
ax.set_ylabel('Component Value')
ax.set_title('Components spread per instance (Iris_Data)');
ax.legend(loc=0);
plt.show()

pca = PCA(n_components=2)
new_iris_data = pca.fit_transform(iris_data)

fig, ax = plt.subplots()
ax.plot(x, new_iris_data[:,0], label = "Component 1")
ax.plot(x, new_iris_data[:,1], label = "Component 2")
ax.set_xlabel('Instances')
ax.set_ylabel('Component Value')
ax.set_title('Components spread per instance (New_Iris_Data)');
ax.legend(loc=0);
plt.show()
```
