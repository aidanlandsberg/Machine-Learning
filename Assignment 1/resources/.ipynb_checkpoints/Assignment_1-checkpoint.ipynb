### 2.2 EigenFaces

#### 2.1.1 Brief Description of EigenFaces

A popular approach to solving the facial recognotion problem is to use PCA to break down the dimensionality into a set of eigenvactors that are able to sucessfully describe the higher dimensional space - in this case facial features. The facial recognotion problem involves investigating an image - represented as an $n$ by $m$ pixel array - for facial features only. In a facial image, (in our case gray scale for simplicity) it should be apparent that the $n$ by $m$ pixel array represents a $n$ by $m$-dimesional space. The PCA then, aims to project the data onto a lower dimensional space in order to determine essential properties. 

#### 2.1.2 Brief History Regarding EigenFaces

Sirovich and Kirby developed the eigenfaces approach for facial recogntion in 1987. By 1991, Mark Turk and Alex Pentland successfully used the eigenfaces approach for facila classification. Sirovich and Kirby initially determined that PCA could be used in oder to obtain a set of basis features from a set of images, namely eigenpictures. This approach, much like that of denoting a sample vector space in terms of linear basis components, reconstructs the original training data by linearly combining the basis components. 

Turk and Pentland then improved the existing system in order to implement automated facial recognition using the eigenfaces. They proposed that the eigenfaces be derived from a covariance matrix of the probibility distribution over the $N$-dimensional vector space that is possible for computational eigen-decomposition of large data sets. Their approach mainly demonstrated the ability to use eigenvectors constructed from arrays of images rather than arrays of pixels.       

#### 2.1.3 Loading the Data Set

```python
from skimage import io
import numpy as np

ic = io.ImageCollection('data/att_faces/*/*.png') # loading all of the faces
fdata = np.array(ic)
# Use the view widget to see the faces
from IPython.html.widgets import interact

def view_image(n=0):
    plt.imshow(ic[n], cmap='gray', interpolation='nearest')
    plt.show()

w = interact(view_image, n=(0, len(ic)-1))
```

As can be seen above, the Olivetti facial data set contains 399 facial images.   

We now set out to flatten the faces array as follows:

```python
n = len(fdata)
d = 92*len(fdata[1,:])
flat_data = fdata.reshape(n,d)

print('Number of faces: \t n = %s' % str(n))
print('Number of features:\t d = %s' % str(d))
print('Flattened data:')
print(flat_data)
```

We can now gather that that the flattened data presents a single matrix where the dimensions correspond to a single pixel representing a single feature across all 399 faces. As shown above, the dimensions of features are far larger than those of the actual faces - making the dimensionality of the problem very large and effectively computationaly inefficient.  

```python
from sklearn.decomposition import PCA

pca_data = PCA(n_components=300) # PCA conducted according to 300 components
PCA_data = pca_data.fit(flat_data)
var = np.array(PCA_data.explained_variance_ratio_)
v_n = len(var)

var_flat = var.reshape(1,v_n)
comp = np.linspace(0.,v_n,v_n)

fig, ax = plt.subplots()
ax.plot(comp, 100*PCA_data.explained_variance_ratio_)
ax.set_xlabel('Components')
ax.set_ylabel('Percentage Variance (%)')
ax.set_title('Variance spread per component');
plt.show()
```

It can cearly be observed from the graphic presented above that the majoity of the variance spread is shown by the approximately first 100 components. Interestingly, after the first 15 components, none of the remaining components yield a variance ratio of greater than 1%. We cna thus conclude that a lower dimensional space may effectively approximate the basis of the facial components. 

```python
# Use the view widget to see the eigenfaces
image_data = np.array(PCA_data.components_)
new_data = image_data.reshape(300,112,92) # resshape the data into an array of 112 x 96 pixel images

from IPython.html.widgets import interact

def view_image(n=0):
    plt.imshow(new_data[n], cmap='gray', interpolation='nearest')
    plt.show()

w = interact(view_image, n=(0, len(new_data)-1))
```

It is evident that the eigenfaces themselves (eigenvalues of the images), describe features that are not necessarily what humans would define - namely eyes, nose, mouths etc. Rather, it shows evident features that can be taken as the basis for recognition much like a linear basis vector would be used to describe a vector counterpart described within the same subspace. The eigenfaces associated witht he highest variance - approximately the first 100 of them - describe an amount of detail that is still recognizable. After the first 100 eigenvectors, the variance is much smaller and the detail shown is far less recognizable, evidently the finer features within the data set.   

```python
pca_data_200 = PCA(n_components=200) # PCA conducted according to 300 components
PCA_data_200 = pca_data_200.fit(flat_data)
var = np.array(PCA_data_200.explained_variance_ratio_)

var_200 = np.array(PCA_data_200.explained_variance_ratio_)
v_200 = len(var_200)

var_flat_200 = var.reshape(1,v_200)
comp_200 = np.linspace(0.,v_200,v_200)

fig, ax = plt.subplots()
ax.plot(comp_200, 100*PCA_data_200.explained_variance_ratio_)
ax.set_xlabel('Components')
ax.set_ylabel('Percentage Variance (%)')
ax.set_title('Variance spread per component');
plt.show()
```

Again it is observed that the majority of the variance is explained by the first 100 components approxiamtely.
