it is evident from loading the iris data set that there exists 90 instances of the iris flower from three separate classes, namely Setosa, Versicolour and Virginica. The data presents the four parameters afterwhich the class it belongs to is shown. 

We can then proceed to fit the model with training data (in this case the aforementioned iris data set) afterwhich dimensionlity reduction can be applied to project the aforemensioned data set onto the first 2 principal components. This is realised using a principal component analysis algorithm that utilises singular value decomposition to project data onto a lower dimensional space according to the maximum variance spread. 

By using the expalined_variance_ratio attribute in the PCA class, it can be deduced that the majority of the variance is indeed explained within the first two principal components - as it should be. The functions output is shown below 

```python
#Conduct PCA and transform the data onto 2 new components 
pca = PCA(n_components=2) 
new_iris_data = pca.fit_transform(iris_data)
#Display Data
print('Variance ratio amongst the first two principal components:')
print(pca.explained_variance_ratio_)
print('Ultametly, it is proven that above 97% of the variance is explained amongst the first 2 components.')
```

The iris data set can be better understood through visualisation of the basic components across each instance.

The second component (a much smaller variance perpindcular to the previously mentioned main component) shows a noticably smaller variance in the data spread - approximately 5%-10%. ### 2.2 The Eigenfaces Approach

#### 2.1.1 Brief Description of EigenFaces

A popular approach to solving the facial recognotion problem is to use PCA to break down the dimensionality into a set of eigenvactors that are able to sucessfully describe the higher dimensional space - in this case facial features. The facial recognotion problem involves investigating an image - represented as an $n$ by $m$ pixel array - for facial features only. In a facial image, (in our case gray scale for simplicity) it should be apparent that the $n$ by $m$ pixel array represents a $n$ by $m$-dimesional space. The PCA then, aims to project the data onto a lower dimensional space in order to determine essential properties. The eigenfaces approach aims to describe a face in terms of a combination of facial features. The PCA then, aims to break down the data set into faces (which are the eigenvectors) that show the maximum variance with regard to the facial features within the relevant data set.

#### 2.1.2 Brief History Regarding EigenFaces

Sirovich and Kirby developed the eigenfaces approach for facial recogntion in 1987. By 1991, Mark Turk and Alex Pentland successfully used the eigenfaces approach for facila classification. Sirovich and Kirby initially determined that PCA could be used in oder to obtain a set of basis features from a set of images, namely eigenpictures. This approach, much like that of denoting a sample vector within a given vector space in terms of linear basis components, reconstructs the original training data by linearly combining the basis components. 

Turk and Pentland then improved the existing system in order to implement automated facial recognition using the eigenfaces. They proposed that the eigenfaces be derived from a covariance matrix of the probibility distribution over the $N$-dimensional vector space that is possible for computational eigen-decomposition of large data sets. Their approach mainly demonstrated the ability to use eigenvectors constructed from arrays of images rather than arrays of pixels. These eigenvectors then, describe the variance with regard to facial features. The advantage of this approach, is that a set of principal components much smaller than the initial dimentionality, can reconstruct a large portion of the images. The eigenvectors representing the majority data spread, can be used to reconstruct faces and be used for a variety of implementations - especially facial recognition. In essence, dimensionality reduction is used to represent the same data from a lower dimensional space. 

#### 2.1.3 Loading the Data Set

```python
from skimage import io
import numpy as np

ic = io.ImageCollection('data/att_faces/*/*.png') # loading all of the faces
fdata = np.array(ic)
# Use the view widget to see the faces
from IPython.html.widgets import interact

def view_image(n=0):
    plt.imshow(ic[n], cmap='gray', interpolation='nearest')
    plt.show()

w = interact(view_image, n=(0, len(ic)-1))
```

As can be seen above, the Olivetti facial data set contains 399 facial images.   

We now set out to flatten the faces array as follows:

```python
n = len(fdata)
d = 92*len(fdata[1,:])
flat_data = fdata.reshape(n,d)

print('Number of faces: \t n = %s' % str(n))
print('Number of features:\t d = %s' % str(d))
print('Flattened data:')
print(flat_data)
```

We can now gather that that the flattened data presents a single matrix where the dimensions correspond to each single pixel representing a single feature across all 399 faces. It is important to note that since the dimensions for each image is identical, each individual pixel should contain some correlation with the identical pixel in the other instances. As shown above, the dimensions of features are fairly large - making the dimensionality of the problem very large and effectively computationaly inefficient. If however, we can find a way to show which pixels particularly (or features) present a large variance accross the data set, we can present these specific features in terms of new components and try to reconstruct our data set in terms of these new components. Ultimately, we are aiming or dimensionality reduction.      

```python
from sklearn.decomposition import PCA

pca_data = PCA(n_components=300) # PCA conducted according to 300 components
PCA_data = pca_data.fit(flat_data)
var = np.array(PCA_data.explained_variance_ratio_)
v_n = len(var)

var_flat = var.reshape(1,v_n)
comp = np.linspace(0.,v_n,v_n)

fig, ax = plt.subplots()
ax.plot(comp, 100*PCA_data.explained_variance_ratio_)
ax.set_xlabel('Components')
ax.set_ylabel('Percentage Variance (%)')
ax.set_title('Variance spread per component');
plt.show()
```

It can cearly be observed from the graphic presented above that the majoity of the variance spread is shown by the approximately first 100 components. Interestingly, after the first 15 components, none of the remaining components yield a variance ratio of greater than 1%. We can thus conclude that these fewer components that describe the maximum variance spread could effectively be used to approximate our data set in terms of a reduced set of components, greatly reducing the dimensionality.  

```python
# Use the view widget to see the eigenfaces
image_data = np.array(PCA_data.components_)
new_data = image_data.reshape(300,112,92) # resshape the data into an array of 112 x 96 pixel images

from IPython.html.widgets import interact

def view_image(n=0):
    plt.imshow(new_data[n], cmap='gray', interpolation='nearest')
    plt.show()

w = interact(view_image, n=(0, len(new_data)-1))
```

It is evident that the eigenfaces themselves (eigenvalues of the images), describe features that are not necessarily what humans would define - namely eyes, nose, mouths etc. - considering all humans have these features, there will exist little variance with regard to these features. Instead, it shows evident features that present a greater variance spread and construct the data set using these projected components for recognition much like a linear basis vector would be used to describe a vector counterpart described within the same subspace. The eigenfaces associated with the highest variance - approximately the first 100 of them - describe an amount of detail that is sufficiently recognizable. After the first 100 eigenvectors, the variance is much smaller and presents the finer features within data set where little variance exists. Theoretically and practically, these features should increase the accuracy but arent essential to sufficiently define/reconstruct the data set.
